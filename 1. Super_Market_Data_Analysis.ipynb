# Super Market Data Analysis

## Overview
This project analyzes supermarket sales data to uncover key insights such as trends, customer preferences, and sales performance across different product categories and locations. The findings can guide decision-making for optimizing operations and marketing strategies.

## Dataset Description
The dataset contains information on:
- **Invoice ID**: Unique ID for each transaction.
- **Branch**: Store branch where the sale occurred.
- **City**: Location of the store.
- **Customer Type**: Membership status (Member/Non-Member).
- **Gender**: Gender of the customer.
- **Product Line**: Category of products sold.
- **Unit Price**: Price of individual items.
- **Quantity**: Number of items purchased.
- **Total**: Total transaction amount.
- **Date**: Date of the sale.
- **Payment**: Mode of payment.
- **Rating**: Customer feedback score.
- **Gross Margin Percentage**: Profit margin for each sale.

## Analysis Goals
1. Identify top-performing branches and products.
2. Understand customer behavior by demographic factors.
3. Explore sales trends over time.
4. Analyze payment preferences and customer ratings.
5. Derive actionable insights for business growth.
