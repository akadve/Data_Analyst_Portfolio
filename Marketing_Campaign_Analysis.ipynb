# Dataset Description

## Overview
The dataset contains information related to marketing campaigns. Here are some initial details:

- **Number of Entries**: The dataset has 2240 entries (rows).
- **Columns**: There are 28 columns:
  - `ID`, `Year_Birth`, `Education`, `Marital_Status`, `Income`, `Kidhome`, `Teenhome`, `Dt_Customer`, `Recency`, `MntWines`, `MntFruits`, `MntMeatProducts`, `MntFishProducts`, `MntSweetProducts`, `MntGoldProds`, `NumDealsPurchases`, `NumWebPurchases`, `NumCatalogPurchases`, `NumStorePurchases`, `NumWebVisitsMonth`, `AcceptedCmp3`, `AcceptedCmp4`, `AcceptedCmp5`, `AcceptedCmp1`, `AcceptedCmp2`, `Complain`, `Z_CostContact`, `Z_Revenue`, `Response`.

## Key Features
- **`ID`**: Unique identifier for each customer.
- **`Year_Birth`**: Year of birth of the customer.
- **`Education`**: Education level of the customer (e.g., Basic, Graduate, etc.).
- **`Marital_Status`**: Marital status of the customer (e.g., Single, Married, etc.).
- **`Income`**: Customer's annual income.
- **`Kidhome`**: Number of children in the household.
- **`Teenhome`**: Number of teenagers in the household.
- **`Dt_Customer`**: Date of customer's enrollment with the company.
- **`Recency`**: Number of days since the last purchase.
- **`MntWines` to `MntGoldProds`**: Monetary spending on different product categories (Wines, Fruits, Meat Products, Fish Products, Sweet Products, Gold Products).
- **`NumDealsPurchases` to `NumStorePurchases`**: Number of purchases made through various channels (Deals, Web, Catalog, Store).
- **`NumWebVisitsMonth`**: Number of web visits in the last month.
- **`AcceptedCmp1` to `AcceptedCmp5`**: Whether the customer accepted the previous campaign offers (1 for yes, 0 for no).
- **`Complain`**: Whether the customer has complained in the last 2 years (1 for yes, 0 for no).
- **`Z_CostContact`**: Cost of contacting the customer.
- **`Z_Revenue`**: Revenue generated from the customer.
- **`Response`**: Response to the current campaign (1 for yes, 0 for no).

## Missing Data
- **`Income`**: Some rows have missing income values, which might need imputation or removal.
- **`Dt_Customer`**: Ensure that dates are properly parsed and handled.
- **Other Columns**: Check for any other potential missing or inconsistent data.


```python
import pandas as pd

# Load the dataset
df = pd.read_csv('marketing_campaign.csv')

# Display the first few rows of the dataframe
print(df.head())

# Get summary statistics of the dataframe
print(df.describe())

# Check the data types of each column
print(df.dtypes)

# Identify missing values
print(df.isnull().sum())
```

# What is the distribution of customer ages?

# Distribution of Customer Ages

## Rationale
Understanding the age distribution of customers can inform age-specific marketing strategies.

## Approach
I've calculated customer ages by subtracting their birth year


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate ages
current_year = 2023
df['Age'] = current_year - df['Year_Birth']

# Plot histogram of customer ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=20, kde=True)
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# How do web, catalog, and store purchases compare?

# Comparison of Purchases by Channel

## Rationale
Comparing different purchasing channels can guide resource allocation and marketing strategy.

## Approach
 summed up the total purchases per channel and visualized the comparison using a stacked bar chart.

## Findings
The stacked bar chart showed that store purchases were significantly higher than web or catalog purchases, suggesting a stronger focus on in-store promotions.



```python
# Sum purchases by channel
channel_purchases = df[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum()

# Plot a stacked bar chart to compare channel-specific purchases
channel_purchases.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Comparison of Purchases by Channel')
plt.xlabel('Channel')
plt.ylabel('Total Number of Purchases')
plt.xticks(rotation=0)
plt.legend(title='Type of Purchase')
plt.tight_layout()
plt.show()
```

# Correlation Between Income and Total Purchases

## Rationale
Investigating correlations between income and purchasing behavior can reveal spending patterns.

## Approach
calculated the total number of purchases per customer and plotted a scatter plot to explore the relationship between income and total purchases.

## Findings
The scatter plot did not show a strong linear correlation, but higher-income customers tended to make slightly more purchases, hinting at some affinity towards spending.


```python
# Assuming 'Total_Purchases' is a sum of all purchase types
df['Total_Purchases'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']

# Scatter plot of Income vs Total Purchases
plt.figure(figsize=(10, 6))
plt.scatter(df['Income'], df['Total_Purchases'], alpha=0.5)
plt.title('Correlation Between Income and Total Purchases')
plt.xlabel('Income')
plt.ylabel('Total Purchases')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Correlation Between Age and Income

plotted a scatter plot with fitted regression line to examine trends between age and income.

# Which education level has the highest average income?

# Average Income by Education Level

## Rationale
Comparing average incomes across education levels can highlight lucrative customer segments.

## Approach
grouped the data by 'Education' and calculated the average income for each group, then visualized these averages using a bar chart.

## Findings
The bar chart showed that customers with a 'Master' degree had the highest average income, followed by 'PhD' holders, highlighting potential high-value segments.



```python
# Group by education level and calculate average income
avg_income_by_education = df.groupby('Education')['Income'].mean().sort_values(ascending=False)

# Plot a bar chart ranking education levels by average income
plt.figure(figsize=(10, 6))
avg_income_by_education.plot(kind='bar')
plt.title('Average Income by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Income')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Does marital status impact purchasing behavior?

# Impact of Marital Status on Purchasing Behavior

## Rationale
Analyzing the effect of marital status on purchases can inform personalized marketing strategies.

## Approach
grouped the data by 'Marital_Status' and calculated the average number of purchases for each type of marital status. A heatmap was used to visualize these correlations.

## Findings
The heatmap indicated that married customers tend to make more purchases through all channels compared to singles, suggesting targeted promotions for married couples could be effective.



```python
# Group by marital status and calculate average purchases
avg_purchases_by_marital = df.groupby('Marital_Status')[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].mean()

# Plot a heatmap to visualize correlations
plt.figure(figsize=(10, 6))
sns.heatmap(avg_purchases_by_marital, annot=True, cmap='viridis')
plt.title('Impact of Marital Status on Purchasing Behavior')
plt.tight_layout()
plt.show()
```

# Do certain product categories have higher average sales than others?

# Average Sales by Product Category

## Rationale
Understanding which product categories have higher average sales can guide inventory management and promotional efforts.

## Approach
Calculated the mean sales for each product category and visualized the results using a bar chart.

## Findings
The bar chart showed that wines and meats had significantly higher average sales compared to fruits, fish products, sweet products, and gold products. This implies that customers prefer alcoholic beverages and meat-based items, suggesting focused promotions could boost sales further.



```python
# Calculate average sales per product category
average_sales_by_category = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean()

# Plot a bar chart comparing average sales across categories
plt.figure(figsize=(10, 6))
average_sales_by_category.plot(kind='bar')
plt.title('Average Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
